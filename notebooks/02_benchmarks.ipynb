# Benchmarks

Reproduce timing tables with the Python API. This mirrors the command-line scripts.

```python
import yaml
from pathlib import Path
from vlp.eval.runner import run_suite

with open('configs/datasets.yaml') as f:
    ds = yaml.safe_load(f)

# Example: Amazon small sweep (adjust fractions as needed)
cfg = dict(dataset='amazon', fractions=[0.002, 0.004, 0.006, 0.008, 0.01], classes=50, iters=1000, repeats=5)
loader = ds[cfg['dataset']]['module'] + '.' + ds[cfg['dataset']]['loader']
out_csv = Path('outputs') / f"{cfg['dataset']}_results.csv"
run_suite(loader, cfg['fractions'], cfg['classes'], cfg['iters'], cfg['repeats'], seed=42, device='auto', dtype='float32', out_csv=out_csv)
print('Wrote', out_csv)
```
