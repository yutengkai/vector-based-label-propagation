# Quickstart: VLP on a toy example

This notebook demonstrates one VLP iteration loop on a small synthetic dataset.

```python
import torch
from vlp.core import vlp

n, d, c = 1000, 64, 50
V = torch.randn(n, d)
Y0 = torch.zeros(n, c)
labels = torch.randint(0, c, (n,))
Y0.scatter_(1, labels[:,None], 1.0)
YT = vlp(V, Y0, iters=10)
YT.shape
```
